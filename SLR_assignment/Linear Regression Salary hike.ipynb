```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.formula.api as smf
from scipy.stats import boxcox
```

## **Model**

```python
df["YearsExperience"].values.reshape(-1,1).shape
```

```python
lin_reg= LinearRegression()
X=df["YearsExperience"].values.reshape(-1,1)
Y=df["Salary"].values.reshape(-1,1)
model=lin_reg.fit(X,Y)
```

```python
X=df["YearsExperience"].values.reshape(-1,1)
Y=df["Salary"].values.reshape(-1,1)
Y
```

# **Cubic Root Transformation**

```python
# Apply the cubic root transformation to the dependent variable (y)
y_cbrt = np.cbrt(Y)

# Fit the model with the cubic root transformed data
lin_reg.fit(X, y_cbrt)
y_pred_cbrt = lin_reg.predict(X)

# Inverse transform the predicted values back to the original scale
y_pred_cbrt_inv = np.power(y_pred_cbrt, 3)

# Calculate the R-squared score for the cubic root transformed data
r2_cbrt = r2_score(Y, y_pred_cbrt_inv)
r2_cbrt
```

```python
plt.plot(X, y_pred_cbrt_inv, color='green', label=f'R-squared (Cubic Root): {r2_cbrt:.3f}')
plt.legend()
plt.title('Delivery Time vs. Sorting Time')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()
```

# **BOX-COX Transformation**

```python
df1=df
df1.describe()
```

```python
X1=df1.iloc[:,:-1].values
Y1=df1.iloc[:,-1].values
```

```python
# Apply the Box-Cox transformation to the dependent variable (y)
y_boxcox, _ = boxcox(Y1)

# Fit the model with the Box-Cox transformed data
lin_reg.fit(X1, y_boxcox)
y_pred_boxcox = lin_reg.predict(X1)

# Inverse transform the predicted values back to the original scale
y_pred_boxcox_inv = np.power(y_pred_boxcox, 1/0.15)

# Calculate the R-squared score for the Box-Cox transformed data
r2_boxcox = r2_score(Y1, y_pred_boxcox_inv)
r2_boxcox

```

```python
plt.plot(X, y_pred_boxcox_inv, color='purple', label=f'R-squared (Box-Cox): {r2_boxcox:.3f}')
plt.legend()
plt.title('Delivery Time vs. Sorting Time')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()
```

# Squared Transformation

```python
# Apply the square transformation to the dependent variable (y)
y_sq = np.square(Y)

# Fit the model with the squared data
lin_reg.fit(X, y_sq)
y_pred_sq =lin_reg.predict(X)

# Inverse transform the predicted values back to the original scale
y_pred_sq_inv = np.sqrt(y_pred_sq)

# Calculate the R-squared score for the squared data
r2_sq = r2_score(Y, y_pred_sq_inv)
r2_sq 
```

```python
plt.plot(X, y_pred_sq_inv, color='orange', label=f'R-squared (Squared): {r2_sq:.3f}')
plt.legend()
plt.title('Delivery Time vs. Sorting Time')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()
```
