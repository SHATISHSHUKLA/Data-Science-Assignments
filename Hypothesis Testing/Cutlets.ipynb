```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
```

```python
df=pd.read_csv("/content/drive/MyDrive/ExcelR/Assignment/Hypothesis testing/Cutlets.csv")
```

```python
df.head()
```

```python
sns.distplot(df["Unit A"])
sns.distplot(df["Unit B"])
plt.legend("Unit A","Unit B")
```

```python
unitA=pd.Series(df.iloc[:,0])
unitB=pd.Series(df.iloc[:,1])
pvalue=stats.ttest_ind(unitA,unitB)
pvalue
```

```python
pvalue[1]
```

pvalue is greater than alpha value hence null hypothesis is accepted
