```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df=pd.read_csv("/content/drive/MyDrive/ExcelR/Assignment/Hypothesis testing/BuyerRatio.csv")
```

```python
df.head()
```

```python
sns.distplot(df['East'])
sns.distplot(df['West'])
sns.distplot(df['North'])
sns.distplot(df['South'])
plt.legend(['East','West','North','South'])
```

```python
fig, plot=plt.subplots()
plot.bar()
```

```python
stats.chi2_contingency([df["East"],df["West"],df["North"],df["South"]])
```

P value= 0.66 which is greater than alpha value so H0 is accepted
